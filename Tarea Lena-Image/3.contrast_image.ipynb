```python
import cv2
import matplotlib.pyplot as plt
import numpy as np


gray_scale_image = cv2.imread("imagenes/gray_image_Lenna.jpg")

img = cv2.imread("imagenes/Lenna.jpg", 0)
ret, binary_img = cv2.threshold(img, 128, 255, cv2.THRESH_BINARY)

# Mostrando la imagen en formato binario
plt.imshow(cv2.cvtColor(binary_img, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# Mostrando el histograma de la imagen en formato binario
plt.hist(binary_img.ravel(), 256, [0, 256], color="black")
plt.show()

mean = np.mean(binary_img)
std_dev = np.std(binary_img)
min_value = np.min(binary_img)
max_value = np.max(binary_img)

print("Mean:", mean)
print("StdDev:", std_dev)
print("Min:", min_value)
print("Max:", max_value)
```

```python
# Mostrando la imagen de la escala de grises con contraste AUMENTADO
alpha = 1.5
beta = 0
gamma = 0

contrast_image = gray_scale_image.copy().astype(np.float32)

cv2.addWeighted(contrast_image, alpha, np.zeros(contrast_image.shape, contrast_image.dtype), 0, beta)
p1, p2 = np.percentile(contrast_image, (50, 50))

contrast_image = np.interp(contrast_image, (p1, p2), (0, 255))
contrast_histogram = contrast_image.copy()

contrast_image = (contrast_image / 255).astype(np.float32)
plt.imshow(contrast_image, cmap='gray', vmin=0, vmax=1)
plt.show()
cv2.imwrite("imagenes/contrast_image_Lenna.jpg", (contrast_image * 255).astype(np.uint8))


```

```python
# Mostrando el histograma de la imagen de la escala de grises con contraste AUMENTADO

plt.hist(contrast_histogram.ravel(), 256, [0, 256], color="black")
plt.show()


mean = np.mean(contrast_histogram)
std_dev = np.std(contrast_histogram)
min_value = np.min(contrast_histogram)
max_value = np.max(contrast_histogram)

print("Mean:", mean)
print("StdDev:", std_dev)
print("Min:", min_value)
print("Max:", max_value)
```
