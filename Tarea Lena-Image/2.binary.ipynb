```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Load the image
img = cv2.imread("imagenes/Lenna.jpg", 0)

# Creating the binary image
ret, binary_img = cv2.threshold(img, 128, 255, cv2.THRESH_BINARY)


plt.imshow(cv2.cvtColor(binary_img, cv2.COLOR_BGR2RGB))
plt.show()
cv2.imwrite("imagenes/binary_image_Lenna.jpg", binary_img)
```
